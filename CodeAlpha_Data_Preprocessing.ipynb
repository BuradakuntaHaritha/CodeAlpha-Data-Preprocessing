## Handle outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data as a NumPy array
data_array = np.random.normal(0, 1, (100, 16))
column_names = ['id', 'rated', 'created_at', 'last_move_at', 'turns', 'victory_status', 'winner', 'increment_code', 'white_id', 'white_rating',
               'black_id', 'black_rating', 'moves', 'opening_eco', 'opening_name', 'opening_ply']

# Convert to DataFrame
data = pd.DataFrame(data_array, columns=column_names)

# Display the first few rows of the dataset
print(data.head())
```

```python
def handle_outliers(df, column, method='remove'):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate IQR (Interquartile Range)
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 -1.5 * IQR
    upper_bound = Q3 +1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    # Handle outliers
    if method == 'remove':
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    elif method == 'cap':
        df[column] = np.clip(df[column], lower_bound, upper_bound)
    
    return df, outliers

# Apply the function to each column
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:  # Process only numeric columns
        data, outliers = handle_outliers(data, column, method='remove')
        print(f"Processed column: {column}")
        print(f"Outliers detected in {column}:")
        print(outliers)
        
        # Visualize the cleaned data
        plt.boxplot(data[column])
        plt.title(f"Boxplot of {column} without Outliers")
        plt.xlabel(column)
        plt.ylabel('Values')
        plt.show()
```

```python
# Load the dataset
dataset=pd.read_csv('train (1).csv')
```

#### Handling Categorical Feature

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```
